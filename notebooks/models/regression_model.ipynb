```
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```
production = np.array([877.952, 1125.295, 1317.822, 2420.78, 2625.742, 2686.193, 2916.534, 3219.187, 3368.42, 3375.228, 3518.441, 3552.19, 3617.581, 3618.559])
consumption = np.array([604.685, 682.931, 957.309, 1625.642, 2446.6, 1745.406, 2445.433, 2898.25, 1875.934, 2157.475, 2394.205, 2060.597, 2096.973, 2419.206])
regr = linear_model.LinearRegression()
regr.fit(production.reshape(-1, 1), consumption)
y = regr.predict(production.reshape(-1, 1))

print('Coeffictients: \n', regr.coef_)
print('Inteercept: \n', regr.intercept_)

## Plot outputs
plt.scatter(production, consumption, color = 'black')
plt.plot(production, y, color = 'blue', linewidth = 3);
```
